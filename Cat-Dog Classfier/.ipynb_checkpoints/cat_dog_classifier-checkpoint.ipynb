```python
# counting no. of cats and dogs pictures
file_names = os.listdir("dogs-vs-cats/train/")
dogCount = 0
catCount = 0
for img_file in file_names:
    name = img_file[0:3]
    if name == 'dog':
        dogCount += 1
    elif name == 'cat':
        catCount += 1

print("Number of dog Images: ", dogCount)
print("Number of cat Images: ", catCount)
```

# Resizing the images to a common size
## here we are storing the resized images into another directory

```python
original_folder = "dogs-vs-cats/train/"
new_folder = "resized_train/"
for i in range(0,25000,10): # we are only taking 2500(1250+1250) images
    img_path = original_folder + file_names[i]
    img = Image.open(img_path)
    img = img.resize((224,224))
    img = img.convert('RGB')

    newImgPath = new_folder + file_names[i]
    img.save(newImgPath)
```

# Creating labels for the images

```python
resized_file_names = os.listdir('resized_train/')
Ylabels = []
for i in range(2500):
    label = resized_file_names[i][0:3]
    if label == 'dog':
        Ylabels.append(1)
    elif label == 'cat':
        Ylabels.append(0)
```

```python
print(resized_file_names[1247:1253], len(resized_file_names))
print(Ylabels[1247:1253], len(Ylabels))
```

```python
# counting the no. of dogs and cats images
values, counts = np.unique(Ylabels, return_counts = True)
print(values, counts)
```

# Converting resized images into numpy arrays

```python
image_directory = 'resized_train/'
image_extension = ['png','jpg']
files = []
[files.extend(glob.glob(image_directory + '*.' + e)) for e in image_extension]
dog_cat_mat = np.asarray([cv2.imread(file) for file in files])
```

```python
print(type(dog_cat_mat), dog_cat_mat.shape)
```

```python
X = dog_cat_mat
Y = np.asarray(Ylabels)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, random_state = 2)
```

```python
print(X_train.shape, Y_train.shape, X_test.shape, Y_test.shape)
```

```python
# Scalling the data
X_train_S = X_train/ 255
X_test_S = X_test / 255
```

# Training the Pretrained Model with our data

```python
tf.__version__, hub.__version__
```

```python
# mobilenet_url = "https://tfhub.dev/google/imagenet/mobilenet_v2_050_224/feature_vector/5"
pretrained_model = tf.keras.applications.MobileNetV2(
    input_shape=(224,224,3),
    alpha=1.0,
    include_top=False,
    weights='imagenet',
    input_tensor=None,
    pooling=None,
    classes=2,
    classifier_activation='softmax'
)
```

```python
# Build the model - Functional API
inputs = tf.keras.Input(shape=(224, 224, 3))
x = pretrained_model(inputs)  # pass symbolic tensor directly — works fine
outputs = tf.keras.layers.Dense(2, activation='softmax')(x)
model = tf.keras.Model(inputs=inputs, outputs=outputs)

# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',  # or 'sparse_categorical_crossentropy'
              metrics=['accuracy'])

model.summary()
```
